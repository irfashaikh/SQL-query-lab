```sql


CREATE or REPLACE TABLE sales_delta (
  order_id INT,
  amount DOUBLE
)
USING DELTA
TBLPROPERTIES (
  delta.enableChangeDataFeed = true
);


ALTER TABLE sales_delta
SET TBLPROPERTIES (
  delta.enableChangeDataFeed = true
);



INSERT INTO sales_delta VALUES
(1, 100),
(2, 200);





UPDATE sales_delta SET amount = 150 WHERE order_id = 1;
DELETE FROM sales_delta WHERE order_id = 2;


SELECT *
FROM table_changes('sales_delta', 0);
-- 0 stands for starting of the delta table         



SELECT *
FROM table_changes('sales_delta', 1, 3);
-- 1 and 3 are the version numbers and to check difference between them
-- table_changes('sales_delta', start_version, end_version)


SELECT *
FROM table_changes('sales_delta', 2) as i
left join table_changes('sales_delta', 4) as j
on i.order_id = j.order_id
where j.order_id is NULL
-- Checking what are the changes were done on the table between version 2 and 4 o a row level.


select max(_commit_version) from table_changes("sales_delta" , 0)
```
