```sql

-- Fetch order details with customer name and product name.

select orders.* , customers.first_name , product.product_name 
from orders join customers
on orders.customer_id = customers.customer_id
join product
on orders.product_id = product.product_id


-- List customers who never placed an order.
select customers.first_name
from customers join orders
on customers.customer_id  = orders.customer_id
where customers.customer_id is null



-- Find customers who placed more than 5 orders.
select count(*) from customers join
(select * , row_number() over(partition by customer_id order by order_amount desc) as rw from orders) as ord
on customers.customer_id = ord.order_id
where ord.rw > 6



-- Show top 5 customers by total spending.
select * , ord.order_amount from customers join
(select * from orders order by order_amount desc ) as ord
on customers.customer_id = ord.customer_id
limit 5
```
