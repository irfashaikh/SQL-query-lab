```sql

-- Find total number of customers per country.
select city , count(customer_id) from customers
group by city


-- Find total orders per customer.
select customers.customer_id , count(*)
from customers join orders 
on customers.customer_id = orders.customer_id
group by customers.customer_id


-- Get average order value per customer
select customers.customer_id , avg(orders.order_amount) 
from customers join orders 
on customers.customer_id = orders.customer_id
group by customers.customer_id


-- Find total order_amount per product category.
select product.category , sum(orders.order_amount) as total_order_amount
from orders join product 
on orders.product_id = product.product_id
group by product.category


```
