```sql

-- Convert all customer names to uppercase.
select upper(customers.first_name) 
from customers

-- Remove extra spaces from full_name.
select trim(customers.first_name)
from customers


-- Find the length of each email address.
select len(email) from customers


-- Extract the first 4 characters of customer names
select customers.first_name
from customers where customers.first_name like '____'


-- Display customer emails in lowercase.
select lower(customers.email) 
from customers


-- Combine first_name and last_name as full_name
select concat(customers.first_name , ' ' , customers.last_name) as full_name
from customers


select replace(customers.email , '@gmail.com' , '@https.com')
from customers 
where customers.email like '%@gmail.com'
```
