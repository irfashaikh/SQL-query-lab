```sql


with emp as (
  select first_name , last_name , email
  from customers
)
select first_name , last_name , email from emp where email like '%@gmail.com'





with final as (
select customers.customer_id , orders.order_amount as avg_final, max(orders.order_amount) as max_final 
from customers join orders
on customers.customer_id = orders.customer_id
group by customers.customer_id
)
select *
from final 
where max_final > (select avg(final.avg_final) from final)



with final as  (
    select orders.* , product.* , (orders.quantity * product.price) as total
    from orders
    join product 
    on orders.product_id = product.product_id
  ),
  final2 as (select final.* , dense_rank() over (partition by customers.city order by final.total desc ) as rw
from customers join final 
on customers.customer_id = final.customer_id
)
select count(*) 
from final2
where rw = 1 or rw = 2








WITH monthly_spend AS (
    SELECT
        customer_id,
        order_date ,
        month(order_date) as curr_month ,
        month(order_date)-1 as pre_month ,
        month(order_date)-2 as pre_pre_month
    FROM orders
    GROUP BY customer_id , order_date
)
SELECT count(*) from 
 monthly_spend
 where 
 month(order_date) = curr_month and month(order_date) -1  = pre_month and month(order_date) -2 = pre_pre_month




select * from customers
select first_name , last_name , customer_id , partition by city from customers 



select t.category , max(t.order_amount) from 
(SELECT orders.* , product.*
FROM ORDERS JOIN PRODUCT 
ON ORDERS.ORDER_ID = PRODUCT.PRODUCT_ID
) t
group by t.category



with final as 
(
  select orders.* , product.* , row_number() over(partition by category order by order_amount desc) as rw 
  from orders join product
  on orders.product_id = product.product_id
)
select category , order_amount from final where rw = 2




with final as 
(
  select orders.* , product.*
  from orders join product
  on orders.product_id = product.product_id
),
final2 as
(
  select final.* , dense_rank() over(partition by category order by order_amount desc) as dr
  from final
)
select category 
from final2
where dr = 10



```
